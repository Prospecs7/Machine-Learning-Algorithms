Using Machine Learning Algorithms on Breast Cancer Data to know the best

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression as lr
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('breast-cancer.csv')
data.head()
```

```python
x = data.iloc[:,2:32]
x.head()
```

```python
encoder = LabelEncoder()
data['EnDiagnosis'] = encoder.fit_transform(data['diagnosis'])
data.head()
```

```python
y = data.EnDiagnosis
y.head()
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 42)
```

```python
kn = KNeighborsClassifier()
kn.fit(x_train, y_train)
```

```python
acc = kn.score(x_train, y_train)
acc
```

```python
pred = kn.predict(x_test)
pred
```

```python
accuracy_score(y_test, pred)
```

```python
confuse = confusion_matrix(y_test, pred)
confuse
```

```python
sv = SVC()
sv.fit(x_train, y_train)
svtrainacc = sv.score(x_train, y_train)
predd = sv.predict(x_test)

svtestacc = accuracy_score(y_test, predd)
svmatric = confusion_matrix(y_test, predd)
```

```python
svmatric
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
dttrainacc = sv.score(x_train, y_train)
pred2 = dt.predict(x_test)

dttestacc = accuracy_score(y_test, pred2)
dtmatrix = confusion_matrix(y_test, pred2)
```

```python
mp = MLPClassifier(max_iter = 250) #best for bigger dataset
mp.fit(x_train, y_train)
mptrainacc = mp.score(x_train, y_train)
pred3 = mp.predict(x_test)

mptestacc = accuracy_score(y_test, pred3)
mpmatrix = confusion_matrix(y_test, pred3)
```

```python
print('Training accuracy for MLP:', mptrainacc)
print('Test accuracy for MLP:', mptestacc)
```

```python
mpmatrix
```

```python
forest = RandomForestClassifier()
forest.fit(x_train, y_train)
foresttrainacc = forest.score(x_train, y_train)
pred4 = forest.predict(x_test)

foresttestacc = accuracy_score(y_test, pred4)
forestmatrix = confusion_matrix(y_test, pred4)
```

```python
print('Training accuracy for forest:', foresttrainacc)
print('Test accuracy for forest:', foresttestacc)
```

```python
forestmatrix
```

```python
gbc = GradientBoostingClassifier()
gbc.fit(x_train, y_train)
gbctrainacc = gbc.score(x_train, y_train)
pred5 = gbc.predict(x_test)

gbctestacc = accuracy_score(y_test, pred5)
gbcmatrix = confusion_matrix(y_test, pred5)
```

```python
print('Training accuracy for forest:', gbctrainacc)
print('Test accuracy for forest:', gbctestacc)
```

```python
lr = lr()
lr.fit(x_train, y_train)
lrtrainacc = lr.score(x_train, y_train)
pred6 = lr.predict(x_test)

lrtestacc = accuracy_score(y_test, pred6)
lrmatrix = confusion_matrix(y_test, pred6)
```

```python
print('Training accuracy for forest:', lrtrainacc)
print('Test accuracy for forest:', lrtestacc)
```
